```python
# Create distribution visualizations for SHS
if shs_merged.shape[0] > 0:
    # Define all months we want to analyze
    lexile_months = [
        "September 2024 Activity Lexile Level",
        "October 2024 Activity Lexile Level",
        "November 2024 Activity Lexile Level",
        "December 2024 Activity Lexile Level",
        "January 2025 Activity Lexile Level",
        "February 2025 Activity Lexile Level",
        "March 2025 Activity Lexile Level",
        "April 2025 Activity Lexile Level",
        "May 2025 Activity Lexile Level"
    ]
    
    # Make sure all data is numeric
    for month_col in lexile_months:
        if month_col in shs_merged.columns:
            shs_merged[month_col] = pd.to_numeric(shs_merged[month_col], errors="coerce")
            print(f"Converted {month_col} to numeric. Records with data: {shs_merged[month_col].notna().sum()}")
        else:
            print(f"WARNING: {month_col} not found in data")
    
    # Make sure strand is extracted
    if "Strand" not in shs_merged.columns:
        def extract_strand(class_str):
            if pd.isna(class_str):
                return "Unknown"
                
            if "STEM" in str(class_str):
                return "STEM"
            elif "ABM" in str(class_str):
                return "ABM"
            elif "HUMSS" in str(class_str):
                return "HUMSS"
            elif "GAS" in str(class_str):
                return "GAS"
            else:
                return "Other"
        shs_merged["Strand"] = shs_merged["Class"].apply(extract_strand)
    
    # Get list of available months in the data
    available_months = [m for m in lexile_months if m in shs_merged.columns]
    
    if not available_months:
        print("ERROR: No lexile data found for any month")
    else:
        # Create visualizations for the most recent months (April and May)
        recent_months = [m for m in ["April 2025 Activity Lexile Level", "May 2025 Activity Lexile Level"] 
                         if m in available_months]
        
        if recent_months:
            plt.figure(figsize=(15, 10 * len(recent_months) // 2))
            
            for i, month_col in enumerate(recent_months):
                month_name = month_col.split()[0]
                
                # Histogram showing distribution by strand
                plt.subplot(len(recent_months), 2, i*2+1)
                for strand in sorted(shs_merged["Strand"].unique()):
                    strand_data = shs_merged[shs_merged["Strand"] == strand][month_col].dropna()
                    if len(strand_data) > 0:
                        sns.histplot(strand_data, alpha=0.6, label=strand, kde=True)
                plt.title(f"{month_name} Lexile Score Distribution by Strand (SHS)")
                plt.xlabel("Lexile Score")
                plt.ylabel("Frequency")
                plt.legend()
                
                # Box plot showing distribution by strand
                plt.subplot(len(recent_months), 2, i*2+2)
                sns.boxplot(x="Strand", y=month_col, data=shs_merged)
                plt.title(f"{month_name} Lexile Score Box Plot by Strand (SHS)")
                plt.xlabel("Strand")
                plt.ylabel("Lexile Score")
                plt.xticks(rotation=45)
            
            plt.tight_layout()
            plt.show()
        
        # Create a comprehensive visualization with all months
        # Determine grid size based on number of months
        n_months = len(available_months)
        n_cols = min(3, n_months)  # Maximum 3 columns
        n_rows = (n_months + n_cols - 1) // n_cols  # Ceiling division
        
        plt.figure(figsize=(6*n_cols, 5*n_rows))
        
        for i, month_col in enumerate(available_months):
            month_name = month_col.split()[0]
            year = month_col.split()[1]
            
            plt.subplot(n_rows, n_cols, i+1)
            sns.boxplot(x="Strand", y=month_col, data=shs_merged)
            plt.title(f"{month_name} {year} Lexile Scores")
            plt.xlabel("Strand")
            plt.ylabel("Lexile Score")
            plt.xticks(rotation=45)
        
        plt.tight_layout()
        plt.suptitle("SHS Lexile Score Distributions by Month", y=1.02, fontsize=16)
        plt.show()
        
        # Create trend visualization
        plt.figure(figsize=(14, 8))
        
        # Prepare data for trend visualization
        trend_data = []
        
        for strand in sorted(shs_merged["Strand"].unique()):
            strand_data = shs_merged[shs_merged["Strand"] == strand]
            
            for month_col in available_months:
                month_name = f"{month_col.split()[0]}"
                mean_lexile = strand_data[month_col].mean()
                if not pd.isna(mean_lexile):
                    trend_data.append({
                        "Strand": strand,
                        "Month": month_name,
                        "Mean Lexile": mean_lexile,
                        "Month_Num": available_months.index(month_col)  # For ordering
                    })
        
        if trend_data:
            trend_df = pd.DataFrame(trend_data)
            
            # Create line plot
            sns.lineplot(
                data=trend_df,
                x="Month_Num",
                y="Mean Lexile",
                hue="Strand",
                markers=True,
                dashes=False,
                style="Strand",
                linewidth=2
            )
            
            # Customize x-axis ticks to show month names
            month_labels = [m.split()[0] for m in available_months]
            plt.xticks(range(len(available_months)), month_labels, rotation=45)
            
            plt.title("SHS Mean Lexile Score Trends by Strand", fontsize=16)
            plt.xlabel("Month", fontsize=12)
            plt.ylabel("Mean Lexile Score", fontsize=12)
            plt.grid(True, alpha=0.3)
            plt.legend(title="Strand", fontsize=10)
            
            plt.tight_layout()
            plt.show()
            
            # Create a heatmap to visualize the progression by strand and month
            plt.figure(figsize=(12, 8))
            
            # Pivot the data for the heatmap
            heatmap_data = trend_df.pivot(index="Strand", columns="Month", values="Mean Lexile")
            
            # Reorder columns to ensure chronological order
            month_order = [m.split()[0] for m in available_months]
            heatmap_data = heatmap_data.reindex(columns=month_order)
            
            # Create the heatmap
            sns.heatmap(
                heatmap_data, 
                annot=True, 
                fmt=".0f", 
                cmap="YlGnBu", 
                linewidths=0.5,
                cbar_kws={"label": "Mean Lexile Score"}
            )
            
            plt.title("SHS Mean Lexile Scores by Strand and Month", fontsize=16)
            plt.tight_layout()
            plt.show()
        
        # Summary statistics for all months
        print("\n=== SHS LEXILE SUMMARY STATISTICS BY STRAND ===")
        
        for month_col in available_months:
            month_name = month_col.split()[0]
            year = month_col.split()[1]
            
            print(f"\nSHS {month_name} {year} Lexile Score Summary Statistics by Strand:")
            month_stats = shs_merged.groupby("Strand")[month_col].agg([
                ('count', 'count'), 
                ('mean', 'mean'), 
                ('std', 'std'), 
                ('min', 'min'), 
                ('25%', lambda x: x.quantile(0.25)),
                ('median', 'median'), 
                ('75%', lambda x: x.quantile(0.75)),
                ('max', 'max')
            ]).round(1)
            
            print(month_stats)
            
            # Overall statistics for the month
            print(f"\nOverall {month_name} {year} Statistics:")
            overall_stats = shs_merged[month_col].agg([
                ('count', 'count'), 
                ('mean', 'mean'), 
                ('std', 'std'), 
                ('min', 'min'), 
                ('25%', lambda x: x.quantile(0.25)),
                ('median', 'median'), 
                ('75%', lambda x: x.quantile(0.75)),
                ('max', 'max')
            ]).round(1)
            
            print(pd.DataFrame(overall_stats).T)  # Transpose to show as a single row
        
        # Month-to-month changes
        print("\n=== MONTH-TO-MONTH LEXILE CHANGES BY STRAND ===")
        
        # Check we have at least two months to compare
        if len(available_months) >= 2:
            for i in range(1, len(available_months)):
                prev_month = available_months[i-1]
                curr_month = available_months[i]
                
                prev_month_name = prev_month.split()[0]
                curr_month_name = curr_month.split()[0]
                
                print(f"\nChange from {prev_month_name} to {curr_month_name}:")
                
                # Calculate changes for each strand
                change_stats = []
                
                for strand in sorted(shs_merged["Strand"].unique()):
                    strand_data = shs_merged[shs_merged["Strand"] == strand]
                    
                    # Get students with data in both months
                    both_months = strand_data.dropna(subset=[prev_month, curr_month])
                    
                    if len(both_months) > 0:
                        # Calculate month-to-month change
                        both_months["change"] = both_months[curr_month] - both_months[prev_month]
                        
                        # Compute statistics
                        change_mean = both_months["change"].mean()
                        change_median = both_months["change"].median()
                        improved = (both_months["change"] > 0).sum()
                        declined = (both_months["change"] < 0).sum()
                        same = (both_months["change"] == 0).sum()
                        
                        change_stats.append({
                            "Strand": strand,
                            "Students": len(both_months),
                            "Mean Change": round(change_mean, 1),
                            "Median Change": round(change_median, 1),
                            "Improved": improved,
                            "Declined": declined,
                            "Same": same,
                            "% Improved": round(improved / len(both_months) * 100, 1)
                        })
                
                # Overall change
                all_both_months = shs_merged.dropna(subset=[prev_month, curr_month])
                if len(all_both_months) > 0:
                    all_both_months["change"] = all_both_months[curr_month] - all_both_months[prev_month]
                    change_mean = all_both_months["change"].mean()
                    change_median = all_both_months["change"].median()
                    improved = (all_both_months["change"] > 0).sum()
                    declined = (all_both_months["change"] < 0).sum()
                    same = (all_both_months["change"] == 0).sum()
                    
                    change_stats.append({
                        "Strand": "OVERALL",
                        "Students": len(all_both_months),
                        "Mean Change": round(change_mean, 1),
                        "Median Change": round(change_median, 1),
                        "Improved": improved,
                        "Declined": declined,
                        "Same": same,
                        "% Improved": round(improved / len(all_both_months) * 100, 1)
                    })
                
                # Display as a DataFrame
                if change_stats:
                    change_df = pd.DataFrame(change_stats)
                    print(change_df.to_string(index=False))
                else:
                    print("No students with data in both months")
        else:
            print("Need at least two months of data to calculate changes")
        
        # Semester analysis
        sem1_months = [m for m in [
            "September 2024 Activity Lexile Level",
            "October 2024 Activity Lexile Level",
            "November 2024 Activity Lexile Level",
            "December 2024 Activity Lexile Level"
        ] if m in available_months]
        
        sem2_months = [m for m in [
            "January 2025 Activity Lexile Level",
            "February 2025 Activity Lexile Level",
            "March 2025 Activity Lexile Level",
            "April 2025 Activity Lexile Level",
            "May 2025 Activity Lexile Level"
        ] if m in available_months]
        
        if sem1_months and sem2_months:
            print("\n=== SEMESTER COMPARISON ===")
            
            # Use the last month of each semester for comparison
            sem1_last = sem1_months[-1]
            sem2_last = sem2_months[-1]
            
            sem1_name = sem1_last.split()[0]
            sem2_name = sem2_last.split()[0]
            
            print(f"\nComparison between {sem1_name} (Semester 1) and {sem2_name} (Semester 2):")
            
            # Calculate changes for each strand
            sem_change_stats = []
            
            for strand in sorted(shs_merged["Strand"].unique()):
                strand_data = shs_merged[shs_merged["Strand"] == strand]
                
                # Get students with data in both semesters
                both_sems = strand_data.dropna(subset=[sem1_last, sem2_last])
                
                if len(both_sems) > 0:
                    # Calculate semester-to-semester change
                    both_sems["change"] = both_sems[sem2_last] - both_sems[sem1_last]
                    
                    # Compute statistics
                    change_mean = both_sems["change"].mean()
                    change_median = both_sems["change"].median()
                    improved = (both_sems["change"] > 0).sum()
                    declined = (both_sems["change"] < 0).sum()
                    same = (both_sems["change"] == 0).sum()

                    sem_change_stats.append({
                        "Strand": strand,
                        "Students": len(both_sems),
                        "Mean Change": round(change_mean, 1),
                        "Median Change": round(change_median, 1),
                        "Improved": improved,
                        "Declined": declined,
                        "Same": same,
                        "% Improved": round(improved / len(both_sems) * 100, 1)
                    })
            
            # Overall change
            all_both_sems = shs_merged.dropna(subset=[sem1_last, sem2_last])
            if len(all_both_sems) > 0:
                all_both_sems["change"] = all_both_sems[sem2_last] - all_both_sems[sem1_last]
                change_mean = all_both_sems["change"].mean()
                change_median = all_both_sems["change"].median()
                improved = (all_both_sems["change"] > 0).sum()
                declined = (all_both_sems["change"] < 0).sum()
                same = (all_both_sems["change"] == 0).sum()
                
                sem_change_stats.append({
                    "Strand": "OVERALL",
                    "Students": len(all_both_sems),
                    "Mean Change": round(change_mean, 1),
                    "Median Change": round(change_median, 1),
                    "Improved": improved,
                    "Declined": declined,
                    "Same": same,
                    "% Improved": round(improved / len(all_both_sems) * 100, 1)
                })
            
            # Display as a DataFrame
            if sem_change_stats:
                sem_change_df = pd.DataFrame(sem_change_stats)
                print(sem_change_df.to_string(index=False))
                
                # Create a visualization of semester improvement
                plt.figure(figsize=(12, 6))
                
                # Bar chart of mean change by strand
                plt.subplot(1, 2, 1)
                sns.barplot(x="Strand", y="Mean Change", data=sem_change_df[sem_change_df["Strand"] != "OVERALL"])
                plt.title(f"Mean Lexile Change from {sem1_name} to {sem2_name}")
                plt.xlabel("Strand")
                plt.ylabel("Mean Lexile Change")
                plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
                plt.xticks(rotation=45)
                
                # Pie chart of overall improvement percentage
                plt.subplot(1, 2, 2)
                overall_row = sem_change_df[sem_change_df["Strand"] == "OVERALL"].iloc[0]
                labels = ['Improved', 'Same', 'Declined']
                sizes = [overall_row['Improved'], overall_row['Same'], overall_row['Declined']]
                colors = ['#5cb85c', '#f0ad4e', '#d9534f']
                
                plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
                plt.axis('equal')
                plt.title(f"Student Outcomes from {sem1_name} to {sem2_name}")
                
                plt.tight_layout()
                plt.show()
            else:
                print("No students with data in both semesters")
        
        # Full year analysis (first month to last month)
        if len(available_months) >= 2:
            first_month = available_months[0]
            last_month = available_months[-1]
            
            first_month_name = first_month.split()[0]
            last_month_name = last_month.split()[0]
            
            print(f"\n=== FULL YEAR ANALYSIS: {first_month_name} to {last_month_name} ===")
            
            # Calculate changes for each strand
            year_change_stats = []
            
            for strand in sorted(shs_merged["Strand"].unique()):
                strand_data = shs_merged[shs_merged["Strand"] == strand]
                
                # Get students with data in both first and last month
                both_months = strand_data.dropna(subset=[first_month, last_month])
                
                if len(both_months) > 0:
                    # Calculate full year change
                    both_months["change"] = both_months[last_month] - both_months[first_month]
                    
                    # Compute statistics
                    change_mean = both_months["change"].mean()
                    change_median = both_months["change"].median()
                    improved = (both_months["change"] > 0).sum()
                    declined = (both_months["change"] < 0).sum()
                    same = (both_months["change"] == 0).sum()
                    
                    # Calculate percentage change
                    both_months["pct_change"] = (both_months[last_month] - both_months[first_month]) / both_months[first_month] * 100
                    pct_change_mean = both_months["pct_change"].mean()
                    
                    year_change_stats.append({
                        "Strand": strand,
                        "Students": len(both_months),
                        "Mean Change": round(change_mean, 1),
                        "Median Change": round(change_median, 1),
                        "Mean % Change": round(pct_change_mean, 1),
                        "Improved": improved,
                        "Declined": declined,
                        "Same": same,
                        "% Improved": round(improved / len(both_months) * 100, 1)
                    })
            
            # Overall change
            all_both_months = shs_merged.dropna(subset=[first_month, last_month])
            if len(all_both_months) > 0:
                all_both_months["change"] = all_both_months[last_month] - all_both_months[first_month]
                change_mean = all_both_months["change"].mean()
                change_median = all_both_months["change"].median()
                improved = (all_both_months["change"] > 0).sum()
                declined = (all_both_months["change"] < 0).sum()
                same = (all_both_months["change"] == 0).sum()
                
                # Calculate percentage change
                all_both_months["pct_change"] = (all_both_months[last_month] - all_both_months[first_month]) / all_both_months[first_month] * 100
                pct_change_mean = all_both_months["pct_change"].mean()
                
                year_change_stats.append({
                    "Strand": "OVERALL",
                    "Students": len(all_both_months),
                    "Mean Change": round(change_mean, 1),
                    "Median Change": round(change_median, 1),
                    "Mean % Change": round(pct_change_mean, 1),
                    "Improved": improved,
                    "Declined": declined,
                    "Same": same,
                    "% Improved": round(improved / len(all_both_months) * 100, 1)
                })
            
            # Display as a DataFrame
            if year_change_stats:
                year_change_df = pd.DataFrame(year_change_stats)
                print(year_change_df.to_string(index=False))
                
                # Create a visualization of full year improvement
                plt.figure(figsize=(15, 10))
                
                # Bar chart of mean change by strand
                plt.subplot(2, 2, 1)
                sns.barplot(x="Strand", y="Mean Change", data=year_change_df[year_change_df["Strand"] != "OVERALL"])
                plt.title(f"Mean Lexile Change from {first_month_name} to {last_month_name}")
                plt.xlabel("Strand")
                plt.ylabel("Mean Lexile Change")
                plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
                plt.xticks(rotation=45)
                
                # Bar chart of percentage change by strand
                plt.subplot(2, 2, 2)
                sns.barplot(x="Strand", y="Mean % Change", data=year_change_df[year_change_df["Strand"] != "OVERALL"])
                plt.title(f"Mean % Lexile Change from {first_month_name} to {last_month_name}")
                plt.xlabel("Strand")
                plt.ylabel("Mean % Change")
                plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
                plt.xticks(rotation=45)
                
                # Pie chart of overall improvement percentage
                plt.subplot(2, 2, 3)
                overall_row = year_change_df[year_change_df["Strand"] == "OVERALL"].iloc[0]
                labels = ['Improved', 'Same', 'Declined']
                sizes = [overall_row['Improved'], overall_row['Same'], overall_row['Declined']]
                colors = ['#5cb85c', '#f0ad4e', '#d9534f']
                
                plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
                plt.axis('equal')
                plt.title(f"Student Outcomes from {first_month_name} to {last_month_name}")
                
                # Bar chart of improvement percentage by strand
                plt.subplot(2, 2, 4)
                sns.barplot(x="Strand", y="% Improved", data=year_change_df[year_change_df["Strand"] != "OVERALL"])
                plt.title(f"Percentage of Students Improved from {first_month_name} to {last_month_name}")
                plt.xlabel("Strand")
                plt.ylabel("% of Students Improved")
                plt.axhline(y=50, color='r', linestyle='--', alpha=0.3)
                plt.xticks(rotation=45)
                
                plt.tight_layout()
                plt.show()
            else:
                print("No students with data in both first and last month")
```
